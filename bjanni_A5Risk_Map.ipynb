```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt
riskDF = pd.DataFrame(columns=['risk', 'likelihood', 'impact', 'score'])
riskDF['risk']=['Denial of service', 'Malware', 'Ransomware', 'Password attacks', 'Phishing', 'Insider threat']
riskDF['likelihood']=[0.4,0.5,0.1,0.8,0.6,0.2]
riskDF['impact']=[4,6,8,6,7,10]
riskDF['score']=riskDF['impact']*riskDF['likelihood']
scoresPivot=riskDF.pivot('impact', 'likelihood', 'score')
print(scoresPivot)
labelsPivot=riskDF.pivot('impact', 'likelihood', 'risk')
print(labelsPivot)
sns.set(rc={'figure.figsize':(16,4)})
p1=sns.heatmap(scoresPivot, cmap="BrBG_r",annot=labelsPivot, fmt='')
```
